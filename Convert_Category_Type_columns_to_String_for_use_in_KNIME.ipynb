```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
for i in tips.columns.to_list():
  if tips[i].dtype == 'category':
    tips[i]=tips[i].astype('string')
  else:
    continue

tips.dtypes
```
